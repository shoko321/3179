```python
new_column_names = ['Country', 'Happiness Score', 'GDP per Capita', 'Social Support', 'Health', 'Freedom',
                   'Government Trust', 'Generosity', 'Year']

df.columns = new_column_names
```

```python
# Define category functions for each column
def categorize_happiness_score(score):
    if score >= 7:
        return "Very High"
    elif score >= 6:
        return "High"
    elif score >= 5:
        return "Medium"
    elif score >= 4:
        return "Low"
    else:
        return "Very Low"

def categorize_gdp(gdp):
    if gdp >= 1.8:
        return "Very High"
    elif gdp >= 1.2:
        return "High"
    elif gdp >= 0.9:
        return "Medium"
    elif gdp >= 0.5:
        return "Low"
    else:
        return "Very Low"

def categorize_social_support(social_support):
    if social_support >= 1.4:
        return "Very High"
    elif social_support >= 1.2:
        return "High"
    elif social_support >= 1.0:
        return "Medium"
    elif social_support >= 0.7:
        return "Low"
    else:
        return "Very Low"

def categorize_health(health):
    if health >= 1.0:
        return "Very High"
    elif health >= 0.8:
        return "High"
    elif health >= 0.6:
        return "Medium"
    elif health >= 0.4:
        return "Low"
    else:
        return "Very Low"

def categorize_freedom(freedom):
    if freedom >= 0.7:
        return "Very High"
    elif freedom >= 0.55:
        return "High"
    elif freedom >= 0.45:
        return "Medium"
    elif freedom >= 0.3:
        return "Low"
    else:
        return "Very Low"

def categorize_government_trust(trust):
    if trust >= 0.2:
        return "Very High"
    elif trust >= 0.165:
        return "High"
    elif trust >= 0.1:
        return "Medium"
    elif trust >= 0.046:
        return "Low"
    else:
        return "Very Low"

def categorize_generosity(generosity):
    if generosity >= 0.3:
        return "Very High"
    elif generosity >= 0.2:
        return "High"
    elif generosity >= 0.17:
        return "Medium"
    elif generosity >= 0.1:
        return "Low"
    else:
        return "Very Low"

# Apply the categorization functions to create new columns
df['Happiness'] = df['Happiness Score'].apply(categorize_happiness_score)
df['GDP per Capita category'] = df['GDP per Capita'].apply(categorize_gdp)
df['Social Support category'] = df['Social Support'].apply(categorize_social_support)
df['Health category'] = df['Health'].apply(categorize_health)
df['Freedom category'] = df['Freedom'].apply(categorize_freedom)
df['Government Trust category'] = df['Government Trust'].apply(categorize_government_trust)
df['Generosity category'] = df['Generosity'].apply(categorize_generosity)

df
```

```python
df.to_csv("world-happiness-report-for-map", index=False)
```
